```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
```

```python
# Read dataset.csv file and convert it into pandas dataframe
df = pd.read_csv("dataset.csv")
```

```python
print(df)
```

```python
# Features
X1 = df["X1"]
X2 = df["X2"]

# Packing Features together
features = list(zip(X1, X2))
```

```python
#Label
label = df["Y=Classification"]
```

```python
# Lets plot these points
plt.plot(X1,X2,"*")
```

```python
# Define Model
model = KNeighborsClassifier(n_neighbors=3, metric="euclidean")
```

```python
# Model training
model.fit(features, label)
```

```python
# Predict classification
predict = model.predict([[4, 7]])
print(predict)
```
